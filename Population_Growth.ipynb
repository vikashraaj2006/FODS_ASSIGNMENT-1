```python
#Scatter Plot
plt.scatter(df['1961'], df['2023'])
plt.xlabel('1961')
plt.ylabel('2023')
plt.title('Scatter Plot of 1961 and 2023')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Create the heatmap with annotations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Example: Creating a histogram for the 'Age' column
plt.hist(df['2023'], bins=10, color='skyblue')
plt.xlabel('2023')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Creating a box plot for the 'GPA' column
sns.boxplot(x='2023', data=df)
plt.title('Box Plot of 2023 year')
plt.show()
```
